```python
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('classic')

import seaborn as sns
sns.set()
```

```python
%matplotlib inline
```

```python
def blurriness(aperture, f, delta, d):
    return - aperture * (f ** 2 * delta / d / (d + delta) + f ** 3 / d / (d + delta))
```

```python
d = -2 # focus distance
f = 50e-3 # focal length
a = 1 / 2.8 # aperture

x = np.linspace(-20, -d - 1e-3, 1000)
y = abs(blurriness(a, f, x, d))

plt.figure(figsize=(8, 6), dpi=150)
plt.xlabel("$\overline{\delta}$ (m)")
plt.ylabel("$|\overline{\Delta}|$ (m)")
plt.xlim(-20, -d)
plt.ylim(0, 1e-3)
plt.axhline(y=3e-4, color='red', linestyle='--', linewidth=1., label="$|\Delta_{\text{lim}}|$")
plt.axvline(x=-4, color='black', linestyle='--', linewidth=1.)
plt.axvline(x=0.8, color='black', linestyle='--', linewidth=1.)
plt.text(-19.8, 3.2e-4, r'$\Delta_{lim}$', fontsize=15)
plt.plot(x, y)
plt.show()
```

```python
d = -2 # focus distance
f = 50e-3 # focal length
a = 1 / 1.8 # aperture

x = np.linspace(-20, -d - 1e-3, 1000)
y1 = abs(blurriness(a, f, x, d))
y2 = abs(blurriness(a / 2, f, x, d))
y3 = abs(blurriness(a / 4, f, x, d))

plt.figure(figsize=(8, 6), dpi=150)
plt.xlabel("$\overline{\delta}$ (m)")
plt.ylabel("$|\overline{\Delta}|$ (m)")
plt.plot(x, y1, label="f/1.8")
plt.plot(x, y2, label="f/2.8")
plt.plot(x, y3, label="f/5.6")
plt.xlim(-20, -d + 1)
plt.ylim(0, 5*1e-4)
plt.legend()
```

```python
d = -5 # focus distance
f = 50e-3 # focal length
a = 1 / 1.8 # aperture

x = np.linspace(-20, -d * 2 - 1e-3, 1000)
y1 = abs(blurriness(a, f, x, d / 2))
y2 = abs(blurriness(a, f, x, d))
y3 = abs(blurriness(a, f, x, d * 2))

plt.figure(figsize=(8, 6), dpi=150)
plt.xlabel("$\overline{\delta}$ (m)")
plt.ylabel("$|\overline{\Delta}|$ (m)")
plt.plot(x, y1, label="2.5m")
plt.plot(x, y2, label="5m")
plt.plot(x, y3, label="10m")
plt.legend()
# plt.axhline(y=1e-4, color='red', linestyle='--', linewidth=1., label="$|\Delta_{\text{lim}}|$")
# plt.text(-19.8, 1.1e-4, r'$\Delta_{lim}$', fontsize=15)
plt.xlim(-20, -d * 2 + 1)
plt.ylim(0, 5*1e-4)
```

```python
d = -5 # focus distance
f = 50e-3 # focal length
a = 1 / 4 # aperture

x = np.linspace(-20, -d - 1e-3, 1000)
y1 = abs(blurriness(a, f / 2, x, d))
y2 = abs(blurriness(a, f, x, d))
y3 = abs(blurriness(a, f * 2, x, d))

plt.figure(figsize=(8, 6), dpi=150)
plt.xlabel("$\overline{\delta}$ (m)")
plt.ylabel("$|\overline{\Delta}|$ (m)")
plt.plot(x, y3, label="100mm")
plt.plot(x, y2, label="50mm")
plt.plot(x, y1, label="25mm")
plt.xlim(-20, -d + 1)
plt.ylim(0, 3*1e-4)
plt.legend()
```

```python
d = -5 # focus distance
f = 50e-3 # focal length
a = 1 / 4 # aperture

x = np.linspace(-20, -d - 1e-3, 1000)
y1 = abs(blurriness(a, f / 2, x, d / 2))
y2 = abs(blurriness(a, f, x, d))
y3 = abs(blurriness(a, f * 2, x, d * 2))

plt.figure(figsize=(8, 6), dpi=150)
plt.xlabel("$\overline{\delta}$ (m)")
plt.ylabel("$|\overline{\Delta}|$ (m)")
plt.plot(x, y3, label="$f'=100 mm$, $\overline{d}=-10 m$")
plt.plot(x, y2, label="$f'=50 mm$, $\overline{d}=-5 m$")
plt.plot(x, y1, label="$f'=25 mm$, $\overline{d}=-2.5m$")
plt.xlim(-20, -d / 2 + 5e-1)
plt.ylim(0, 2 * 1e-4)
plt.legend()
```
